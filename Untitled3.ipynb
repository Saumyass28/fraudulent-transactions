```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import xgboost as xgb
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, roc_curve, precision_recall_curve
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'C:/Users/ASUS/Downloads/Fraud.csv'
df = pd.read_csv(file_path)

# Drop columns that are not useful for the model
df = df.drop(columns=['step', 'nameOrig', 'nameDest'])

# Features and target
X = df.drop('isFraud', axis=1)
y = df['isFraud']

# Define the column transformer for encoding categorical features
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(), ['type'])
    ],
    remainder='passthrough'  # Keep the remaining columns (numeric) as they are
)

# Create a pipeline with preprocessing and XGBoost model
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss'))
])

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

```

```python
# Train the model
pipeline.fit(X_train, y_train)
```

```python
# Make predictions
y_pred = pipeline.predict(X_test)
y_prob = pipeline.predict_proba(X_test)[:, 1]
```

**3. Variable Selection**

```python
#import matplotlib.pyplot as plt
import xgboost as xgb

# Make sure the pipeline has been fitted
pipeline.fit(X_train, y_train)

# Access the trained XGBoost model from the pipeline
xgb_model = pipeline.named_steps['classifier']

# Plot feature importances
xgb.plot_importance(xgb_model)
plt.show()
```

**4. Model Performance Evaluation**

```python
# Evaluate the model
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
# ROC-AUC
roc_auc = roc_auc_score(y_test, y_prob)
print("ROC-AUC:", roc_auc)
```

```python
# ROC Curve
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
plt.plot(fpr, tpr, label=f'XGBoost AUC = {roc_auc:.2f}')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='best')
plt.show()
```

```python
# Precision-Recall Curve
precision, recall, _ = precision_recall_curve(y_test, y_prob)
plt.plot(recall, precision, label="Precision-Recall Curve")
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend(loc='best')
plt.show()
```

**5. Key Factors Predicting Fraudulent Transactions**

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), ['amount', 'oldbalanceOrg', 'newbalanceOrig', 'oldbalanceDest', 'newbalanceDest']),
        ('cat', OneHotEncoder(), ['type'])
    ])

pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss'))
])

# Fit pipeline
pipeline.fit(X_train, y_train)

# Extracting feature names after transformation
feature_names = pipeline.named_steps['preprocessor'].transformers_[0][1].get_feature_names_out()  # For numeric features
categorical_feature_names = pipeline.named_steps['preprocessor'].transformers_[1][1].get_feature_names_out()  # For categorical features
all_feature_names = list(feature_names) + list(categorical_feature_names)

# Extracting feature importances 
xgb_model = pipeline.named_steps['classifier']
importances = xgb_model.feature_importances_

if len(importances) == len(all_feature_names):
    feature_importances = pd.Series(importances, index=all_feature_names).sort_values(ascending=False)
    print(feature_importances)
else:
    print(f"Mismatch: {len(importances)} importances, {len(all_feature_names)} feature names.")
```
